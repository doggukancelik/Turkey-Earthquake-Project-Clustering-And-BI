```python
from geopy.geocoders import Nominatim
lat = []
lon = []
geolocator = Nominatim(user_agent='myapplication')
for index, row in df.iterrows():
    lat.append(geolocator.geocode(df['YER'][index]).latitude)
    lon.append(geolocator.geocode(df['YER'][index]).longitude)
```

```python
df["lat"] = lat
df["lon"] = lon
```
